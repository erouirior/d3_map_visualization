```python
import folium
from folium.plugins import MarkerCluster
SF_COORDINATES=[37.7749,-122.4194]
color1='LightPink'
color2='LightSeaGreen'
color3='LightSalmon'
```

```python
map_brew = folium.Map(location=SF_COORDINATES, zoom_start=12,tiles='Cartodb Positron')
```

```python
for index, row in data.iterrows():
    map_brew.circle_marker([row['latitude'],row['longitude']],radius=0,fill_color=color1,line_color=color1)
```

```python
map_brew.save('lists.html')
```

```python
for index, row in data.iterrows():
    if row['room_type']=='Entire home/apt':
        map_brew.circle_marker([row['latitude'],row['longitude']],radius=0,fill_color=color1,line_color=color1)
    elif row['room_type']=='Private room':
        map_brew.circle_marker([row['latitude'],row['longitude']],radius=0,fill_color=color2,line_color=color2)
    else:
        map_brew.circle_marker([row['latitude'],row['longitude']],radius=0,fill_color=color3,line_color=color3)
```

```python
map_brew.save('map_roomtype.html')
```

```python
# definition of the boundaries in the map
district_geo = r'neighbourhoods.geojson'
```

```python
neigh=pd.read_csv('neighbourhoods.csv')
```

```python
neigh.head(2)
```

```python
data.groupby('neighbourhood').size()[:5]
```

```python
# calculating total number of incidents per district
avai=data[['neighbourhood','pricePBedRoom']].groupby('neighbourhood').median()
avai = pd.DataFrame(avai.reset_index())
#gerritsen_Beach=np.array(['Gerritsen Beach',0])
#piv=np.array(['Port Ivory',0])
#baysi=np.array(['Bay Terrace, Staten Island',0])
#avai.loc[202]=gerritsen_Beach
#avai.loc[203]=piv            
#avai.loc[204]=baysi
```

```python
newavai=pd.merge(avai,neigh,on='neighbourhood',how='outer')
```

```python
newavai.to_json('neigh_price.json')
newavai.columns = ['neighbourhood', 'pricePBedRoom','neighbourhood_group']
```

```python
newavai.price=newavai.pricePBedRoom.apply(pd.to_numeric)
```
